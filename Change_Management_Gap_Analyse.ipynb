```python
# 1. Setup
!pip install -U sentence-transformers pandas sklearn

import pandas as pd
from sentence_transformers import SentenceTransformer, util
from sklearn.cluster import KMeans
import torch

# 2. Modell & Daten laden
model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

# Hier deine Dateien hochladen (meine_keywords.csv & wettbewerber_keywords.csv)
def load_data(file):
    try:
        df = pd.read_csv(file, sep=None, engine='python', encoding='utf-8-sig')
        # Sucht die Keyword-Spalte automatisch
        kw_col = [c for c in df.columns if 'keyword' in c.lower() or 'suchbegriff' in c.lower()][0]
        return df[kw_col].dropna().unique().tolist()
    except:
        return []

meine_kws = load_data('meine_keywords.csv')
comp_kws = load_data('wettbewerber_keywords.csv')

# 3. Embeddings & Berechnung
print(f"Analyse startet für Thema: Change Management")
emb_meine = model.encode(meine_kws, convert_to_tensor=True)
emb_comp = model.encode(comp_kws, convert_to_tensor=True)

results = []
for i, kw in enumerate(comp_kws):
    # Falls du noch gar keine eigenen KWs hast, ist die Similarity 0
    score = torch.max(util.cos_sim(emb_comp[i], emb_meine)).item() if meine_kws else 0
    results.append({'Keyword': kw, 'Match_Score': score})

df_res = pd.DataFrame(results).sort_values(by='Match_Score')

# 4. Bonus: Clustering der Wettbewerber-Keywords (Themenwelten finden)
# Wir gruppieren die Konkurrenz-Keywords in 5 Themenbereiche
num_clusters = min(5, len(comp_kws))
clusters = KMeans(n_clusters=num_clusters, random_state=42).fit_predict(emb_comp.cpu())
df_res['Themen_Gruppe'] = clusters

# 5. Ausgabe
print("\n--- DEINE TOP 15 STRATEGISCHEN LÜCKEN (Nach Themen sortiert) ---")
print(df_res.head(15)[['Themen_Gruppe', 'Keyword', 'Match_Score']])

df_res.to_csv('change_management_strategie.csv', sep=';', index=False)
```
